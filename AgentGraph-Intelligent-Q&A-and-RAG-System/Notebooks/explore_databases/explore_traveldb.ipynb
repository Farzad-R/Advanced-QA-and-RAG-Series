```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pyprojroot import here
```

```python
# Connect to SQLite database
db_path = here('data/travel.sqlite')
conn = sqlite3.connect(db_path)
```

```python
# Get list of all tables
query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql(query, conn)
tables
```

```python

# Perform EDA for each table
for table in tables['name']:
    print(f"\n\n-- Table: {table} --")
    
    # Load data from the table into a DataFrame
    df = pd.read_sql(f"SELECT * FROM {table}", conn)
    
    # Show the first few rows
    print(f"First 5 rows of {table}:\n", df.head())
    
    # Show the schema (column names and data types)
    print("\nColumn Data Types:")
    print(df.dtypes)
    
    # Show basic statistics
    print("\nBasic statistics:")
    print(df.describe(include='all'))
    
    # Check for missing values
    print("\nMissing values in each column:")
    print(df.isnull().sum())

    # # Check correlations (only for numeric columns)
    # if df.select_dtypes(include='number').shape[1] > 1:
    #     print("\nCorrelation matrix:")
    #     corr = df.corr()
    #     print(corr)
        
        # # Plot correlation heatmap
        # plt.figure(figsize=(10, 6))
        # sns.heatmap(corr, annot=True, cmap="coolwarm", fmt='.2f')
        # plt.title(f"Correlation Heatmap for {table}")
        # plt.show()
    
    # Plot histograms for numeric columns
    num_cols = df.select_dtypes(include='number').columns
    if len(num_cols) > 0:
        df[num_cols].hist(figsize=(10, 8), bins=20)
        plt.suptitle(f"Histograms of numeric columns in {table}")
        plt.show()
    
    # Plot bar charts for categorical columns (if less than 20 unique values)
    cat_cols = df.select_dtypes(include='object').columns
    for col in cat_cols:
        if df[col].nunique() < 20:
            plt.figure(figsize=(8, 4))
            df[col].value_counts().plot(kind='bar')
            plt.title(f"Bar chart of {col} in {table}")
            plt.ylabel('Frequency')
            plt.show()

# Close the database connection
conn.close()
```
